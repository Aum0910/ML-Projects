```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, accuracy_score
```

```python
target = 'default'
```

```python
df.columns
```

```python
cat_cols = ['SEX', 'EDUCATION', 'MARRIAGE']
num_cols = ['LIMIT_BAL', 'AGE']
```

```python
X = df[cat_cols + num_cols]
y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y)
```

```python
X_train.shape
```

```python
cat_transformer = OneHotEncoder(drop = 'first', handle_unknown = 'ignore')
```

```python
num_transformer = StandardScaler()
```

```python
preprocessor = ColumnTransformer(transformers = [('cat', cat_transformer, cat_cols), ('num', num_transformer, num_cols)])
```

```python
log_reg = LogisticRegression(max_iter = 1000, solver = 'liblinear')
```

```python
pipeline = Pipeline(steps = [('preprocess', preprocessor), ('model', log_reg)])
```

```python
cv = StratifiedKFold(n_splits = 5, shuffle = True)
```

```python
roc_auc_cv = cross_val_score(pipeline, X_train, y_train, cv = cv, scoring = 'roc_auc')
```

```python
print("CV ROC AUC Scores: ", roc_auc_cv)
```

```python
roc_auc_cv.mean()
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_test_pred = pipeline.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("Test Accuracy:", test_accuracy)
```

```python
mapping = {-2 : "Duly paid", -1 : "Duly paid", 0 : "Duly paid", 1 : "One month delay", 2 : "Two month delay", 3 : "Three month delay", 4 : "Four month", 5 : "Five month", 6 : "Six month", 7 : "Seven month", 8 : "Eight month"}
pay_cols = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
df[pay_cols] = df[pay_cols].replace(mapping)
```

```python
df['PAY_2'].value_counts()
```

```python
cat_cols = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0']
```

```python
num_cols = ['LIMIT_BAL','AGE','BILL_AMT1','PAY_AMT1']
```

```python
X = df[cat_cols + num_cols]
y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y)
```

```python
X_train_cat = pd.get_dummies(X_train[cat_cols], drop_first = True)
X_test_cat = pd.get_dummies(X_test[cat_cols], drop_first = True)
```

```python
X_train_cat, X_test_cat = X_train_cat.align(X_test_cat, join = 'left', axis = 1, fill_value = 0)
```

```python
X_train_cat.columns
```

```python
scaler = StandardScaler()
```

```python
X_train_num = scaler.fit_transform(X_train[num_cols])
X_test_num = scaler.transform(X_test[num_cols])
```

```python
X_train_num = pd.DataFrame(X_train_num, columns = num_cols, index = X_train.index)
```

```python
X_train_final = pd.concat([X_train_cat, X_train_num], axis = 1)
```

```python
X_test_num = pd.DataFrame(X_test_num, columns = num_cols, index = X_test.index)
```

```python
X_test_final = pd.concat([X_test_cat, X_test_num], axis = 1)
```

```python
log_reg = LogisticRegression(max_iter = 1000, solver = 'liblinear')
log_reg.fit(X_train_final, y_train)
```

```python
log_reg.get_params()
```

```python
y_test_pred = log_reg.predict(X_test_final)
```
